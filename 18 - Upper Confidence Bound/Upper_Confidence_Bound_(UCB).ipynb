# Upper_Confidence_Bound_(UCB)

### Importing Libraries And Datasets

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_set = pd.read_csv('Ads_CTR_Optimisation.csv')
data_set
```

### Implementing UCB

```python
import math
N = 10000
ADS = 10
ads_selected = []
no_of_times_selected = [0] * ADS
sum_of_rewards = [0] * ADS
total_reward = 0
for _ in range(N):
    ad = 0
    mx_upper_bound = 0
    for i in range(ADS):
        if no_of_times_selected[i] > 0:
            average_reward = sum_of_rewards[i] / no_of_times_selected[i]
            delta_i = math.sqrt(3/2 * math.log(_ + 1) / no_of_times_selected[i])
            upper_bound = average_reward + delta_i
        else:
            upper_bound = 1e400
        if upper_bound > mx_upper_bound:
            mx_upper_bound = upper_bound
            ad = i
    ads_selected.append(ad)
    no_of_times_selected[ad] = no_of_times_selected[ad] + 1
    reward = data_set.values[_, ad]
    sum_of_rewards[ad] = sum_of_rewards[ad] + reward
    total_reward += data_set.values[_, ad]
```

```python
print(ads_selected, no_of_times_selected, sum_of_rewards, total_reward)
```

### Visualizing Data

```python
plt.hist(ads_selected)
plt.title("Histogram of ADs Selected")
plt.xlabel('ADs')
plt.ylabel('No. of Times Selected')
plt.show()
```
