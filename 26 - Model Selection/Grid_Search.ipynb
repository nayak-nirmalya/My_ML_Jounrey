# Kernel SVM And Grid Search

### Importing Datasets And Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_set = pd.read_csv('Social_Network_Ads.csv')
print(data_set)
print(data_set.describe())
```

```python
X = data_set.iloc[:, 3:5].values
X
```

```python
y = data_set.iloc[:, -1].values
y
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
print(X)
```

### Spliting Datasets

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
print(X_train, X_test, y_train, y_test)
```

### Clasifying

```python
from sklearn.svm import SVC
```

```python
classifier = SVC(kernel = 'sigmoid')
```

```python
classifier.fit(X_train, y_train)
```

### Predicting

```python
y_pred = classifier.predict(X_test)
```

```python
for _ in zip(y_pred, y_test):
    print(*_)
```

### Confussion Matrix

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
print(cm)
```

```python
cm.shape
```

### Applying K_Fold_CrossValidation

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier, X = X_train, y = y_train, cv = 10, n_jobs = -1)
print(accuracies.mean(), accuracies.std())
```

### Applying Grid Search To Find Best Model And Best Parameters

```python
from sklearn.model_selection import GridSearchCV
parameters = [
    {'C' : [1, 10, 100, 1000], "kernel" : ['linear']},
    {'C' : [1, 10, 100, 1000], "kernel" : ['rbf'], 'gamma' : [0.1, 0.5, 0.01, 0.001, 0.0001]}
]
grid_search = GridSearchCV(estimator = classifier, param_grid = parameters, scoring = 'accuracy', cv = 10, n_jobs = -1)
grid_search = grid_search.fit(X_train, y_train)
```

```python
best_accuracy = grid_search.best_score_
print(best_accuracy)
best_param = grid_search.best_params_
print(best_param)
```
