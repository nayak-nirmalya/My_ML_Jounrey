# Thompson Sampling

# Importing Libraries And Datasets

```python
import numpy
import pandas
import matplotlib.pyplot as plt
```

```python
data_set = pandas.read_csv('Ads_CTR_Optimisation.csv')
data_set
```

```python
for i in range(len(data_set)):
    print(data_set.values[i])
```

### Implementing Algorithm

```python
import math, random
N = 10000
ADS = 10
ads_selected = []
no_of_rewards = [0] * ADS
no_rewards = [0] * ADS
total_reward = 0
for _ in range(N):
    ad = 0
    max_random = 0
    for i in range(ADS):
        random_beta = random.betavariate(no_of_rewards[i] + 1, no_rewards[i] + 1)
        if random_beta > max_random:
            max_random = random_beta
            ad = i
    ads_selected.append(ad)
    reward = data_set.values[_, ad]
    if reward == 1:
        no_of_rewards[ad] += 1
    else:
        no_rewards[ad] += 1
    total_reward += reward
```

```python
print(ads_selected, no_of_rewards, no_rewards, total_reward)
```

### Visualizing Data

```python
plt.hist(ads_selected)
plt.title("Histogram of ADs Selected")
plt.xlabel('ADs')
plt.ylabel('No. of Times Selected')
plt.show()
```
